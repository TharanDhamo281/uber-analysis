```python
import pandas as pd 
df=pd.read_csv("taxi.csv")
df.tail()
```

```python
x=df.drop('Numberofweeklyriders',axis=1)
y=df['Numberofweeklyriders']
```

```python

#using sklearn train test
from sklearn.model_selection import train_test_split
""" 
You're importing a function named train_test_split from the Scikit-learn (sklearn) library.
train_test_split is used to split your dataset into two parts:
Training set: used to train the model
Testing set: used to test the model
"""



x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.25,random_state=0)
""" 

This line splits your data into training and testing sets.

Letâ€™s break it down:

x: Your features (input data). For example: height, weight, age, etc.

y: Your labels/target (output you want to predict). For example: healthy or not.

x_train: The part of x used for training the model.

x_test: The part of x used for testing the model.

y_train: The part of y used with x_train to teach the model.

y_test: The part of y used with x_test to evaluate the model.

ðŸ“Œ Parameters inside the function:
test_size=0.25:

25% of the data will go to the test set

75% will go to the train set

random_state=0:

This ensures the split is the same every time you run the code.

Useful for reproducibility

"""
```

```python

#model TTraining 
from sklearn.linear_model import LinearRegression

""" Linear regression is used to find the relationship between input (x) and output (y) using a straight line"""
model=LinearRegression()
""" This creates an instance of the linear regression model and stores it in the variable model.

Now model is ready to be trained"""
model.fit(x_train,y_train)
""" This line trains the model using your training data (x_train and y_train).

.fit() is the method used to make the model learn the pattern or relationship between the input (features) and output (labels).

After this step, the model will be ready to make predictions on new data."""
```

```python

print("Training Score accuracy:- ",model.score(x_train,y_train))
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
```
